# **Drawing Plots**

```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
%matplotlib inline

data = pd.read_csv("blood-pressure.csv")
```

### **Barplot**

```python

sn.barplot(x = "Index", y = "Age", hue="Blood_Pressure", data = data)
```

## **Histogram**

```python
plt.hist(data['Blood_Pressure'], bins = 20)
```

### **Distribution or Density Plot**

```python
sn.distplot(data['Blood_Pressure'])
```

### **Box Plot**

```python
box = sn.boxplot(data['Blood_Pressure'])
```

```python
data[data['Age'] > 50][['Index', 'Age', 'Blood_Pressure']]
```

### **Comparing Distributions**

```python
sn.distplot(data[data['Age'] == 47]
            ['Blood_Pressure'],
            color='r',
            label= "BP of age = 50")
sn.distplot(data[data['Age'] == 67]
            ['Blood_Pressure'],
            color='y',
            label="BP of age = 67");
```

### **Scatter Plot**

```python
blood_df = data[data['Blood_Pressure'] == 150]
plt.scatter(x = blood_df.Index, y = blood_df['Age'])
```

```python
sn.regplot(x = 'Age', y = 'Blood_Pressure', data = blood_df)
```

### **Pair Plot**

```python
features = ['Index', 'Age', 'Blood_Pressure']
sn.pairplot(data[features], size = 2)
```

### **Correlation and Heatmap**

```python
data[features].corr()
```

```python
sn.heatmap(data[features].corr(), annot = True)
```
